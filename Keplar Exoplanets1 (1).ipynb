```python
df = df.drop(columns=["kepler_name","koi_teq_err1","koi_teq_err2"])
```

```python
df = df.dropna()
df = df.reset_index(drop=True)
df.shape
```

```python
df["koi_fpflag_nt"].value_counts()
```

```python
df = df[df["koi_fpflag_nt"] <= 1]
df["koi_fpflag_nt"].value_counts()
```

```python
df["koi_disposition"].value_counts()
```

```python
sns.countplot(df['koi_disposition'],palette='winter_r')
```

```python
def disposition_encode(x):
    if x == "FALSE POSITIVE":
        return -1
    if x == "CANDIDATE":
        return 0
    if x == "CONFIRMED":
        return 1
df["koi_disposition"] = df["koi_disposition"].apply(lambda x: disposition_encode(x))

df["koi_disposition"].value_counts()
```

```python
df["koi_pdisposition"].value_counts()
```

```python
sns.countplot(df['koi_pdisposition'],palette='winter_r')
```

```python
df["koi_pdisposition"] = df["koi_pdisposition"].apply(lambda x : 1 if x == "CANDIDATE" else -1)
df["koi_pdisposition"].value_counts()
```

```python
import matplotlib.pyplot as plt

Candidate = df[df.koi_pdisposition == 1]
FP = df[df.koi_pdisposition == -1]

raf, decf = FP.ra, FP.dec
rac, decc = Candidate.ra, Candidate.dec

fig = plt.figure(figsize=(8,8))

plt.scatter(raf, decf, s=4, label='False Positive')
plt.scatter(rac, decc, s=4, label="Candidates")

plt.xlabel("Right Ascension")
plt.ylabel("Declination")

plt.legend()
```
